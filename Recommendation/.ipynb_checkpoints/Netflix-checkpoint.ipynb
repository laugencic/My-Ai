7) Find all titles released in the year 2020

```python
year=data[data['Release Year']==2020]['title']
print(year)
#953 movies were realesed in 2020
```

8)Find all content added to Netflix in January 2021

```python
jan=data[(data['Date Added'].dt.year==2021) & (data['Date Added'].dt.month==1)]
print(jan.value_counts())
```

9)Find movies that are longer than 120 minutes

```python
long=data[(data['type']=="Movies") & (data['duration']>120)]
print(long)
```
