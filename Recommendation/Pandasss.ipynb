9)Find movies that are longer than 120 minutes

```python
#creating a new column of only movies since the duration nwas mixed with seasons and minutes
data['duration'] = data['duration'].str.replace(' min', '', regex=False)
data['duration'] = pd.to_numeric(data['duration'], errors='coerce')

data['Movie Duration']=np.where(
    data['type']=="Movie",
    data['duration'],
    np.nan
)

long=data[data['Movie Duration']>120]
print(long)
```
