9)Find movies that are longer than 120 minutes

```python
data['duration_value']=data['duration'].str.extract(r'(\d+)').astype(float)
#this extracts the numeric value of the shows and movies duration then converts them to a float datatype
# (\d+) >>> \d >> digits from 0 to 9, +>> one or more 

data['movie duration']=np.where(
    data['type']=="Movie",
    data['duration_value'],
    np.nan
)
data['show seasons']=np.where(
    data['type']=="TV Show",
    data['duration_value'],
    np.nan
)
long=data[data['movie duration']>120][['title','movie duration']]
print(long)
```

10)Find TV shows with more than 3 seasons

```python
long=data[data['show seasons']>3][['title','show seasons']]
print(long)
```

13)Which rating (PG-13, TV-MA, etc.) is most common for movies? For TV shows

```python
movie_ratings=data[data['type']=="Movie"]['rating'].value_counts().head(5)
print("Movies",movie_ratings)

shows_ratings=data[data['type']=="TV Show"]['rating'].value_counts().head(5)
print("Shows",show_ratings)

#Tv-ma is the most common for movies and tv shows meaning majority of the content is tailored for the older audiences 
```

14). What's the average duration of movies by year (line chart showing trend)?

```python
ave_duration=data.groupby('release_year')['movie duration'].mean().sort_index()
ave_duration.plot()
plt.show()
```
